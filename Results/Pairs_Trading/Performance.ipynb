# Performance Technical Strategy

## Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Get data Reports

```python


# URL ของ Google Sheets
url = "https://docs.google.com/spreadsheets/d/1M2J41NhgnGg0SxwIskfqQRSHmYGtRlT6ztNaQm64lsE/export?format=csv"

# อ่านไฟล์ CSV
df = pd.read_csv(
    'Reports - Pair Trading.csv')

# แสดงข้อมูล
df
```

### Group by Name strategy

```python
data = df.copy()
```

```python
strategy_name = data.groupby('Strategy' , sort=False)
```

```python
strategy_name
```

```python
strategies = []
for group_name, group_data in strategy_name:
    strategies.append(group_name)
   
```

```python
strategies
```

```python
title = 'Pair Trading With Kalman Filter [2007 - 2023]'
```

## Profit Factor

```python
Evaluation = 'Profit Factor'
```

```python
df[Evaluation] = df[Evaluation].fillna(0).round(2)
```

```python
Profit_Factor_h4 = df.loc[df['Time frame'] == 'H4', Evaluation]
Profit_Factor_h1 = df.loc[df['Time frame'] == 'H1', Evaluation]
Profit_Factor_d1 = df.loc[df['Time frame'] == 'D1', Evaluation]
Profit_Factor_m30 = df.loc[df['Time frame'] == 'M30', Evaluation]
```

```python
x = np.arange(len(strategies))
timeframe = {
    'D1': Profit_Factor_d1,
    'H4': Profit_Factor_h4,
    'H1': Profit_Factor_h1,
    'M30': Profit_Factor_m30,
}
width = 0.2  # the width of the bars
multiplier = 0


fig, ax = plt.subplots(layout='constrained' , figsize=(48,12))

for TF, measurement in timeframe.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement , width, label= TF )
    ax.bar_label(rects, padding=3, fontsize=20)
    multiplier += 1


# Add horizontal line at y=50
# ax.axhline(y=1, color='r', linestyle='--', linewidth=2)
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_title(f'{title}', fontsize=40)
ax.set_ylabel('Profit Factor', fontsize=30)
ax.set_xticks(x + width, strategies, fontsize=30 )
ax.legend(loc='upper left', ncols=3, fontsize=30)
# Set y-axis tick label fontsize here
ax.tick_params(axis='y', labelsize= 20)
ax.set_ylim(0, 2)

plt.show()
    
```
