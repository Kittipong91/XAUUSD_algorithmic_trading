```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
data_init = pd.read_csv('../Data_Forex/EURUSD/EURUSD_H4 - EURUSD_H4.csv',parse_dates=True,index_col=0)
data_init
```

```python
data_exness = pd.read_csv('../Data_Forex/EURUSD/EURUSD_H4_Exness.csv',parse_dates=True,index_col=0)
data_exness
```

```python
data_exness.info()
```

```python
data_init.info()
```

```python
startDate = '2018-07-10 08:00:00'
endDate = '2023-03-10 16:00:00'
```

```python
data_exness = data_exness[startDate: endDate]
data_init = data_init[startDate : endDate]
```

```python
data_exness.info()
```

```python
data_init.info()
```

```python
data_exness['spread'] = data_exness['Close'] - data_init['Close']
```

```python
fig, axes = plt.subplots(2,1, figsize = (26,12), gridspec_kw = {'height_ratios':[2,1]})
plt.subplots_adjust(hspace=0)

axes[0].plot(data_exness['Close'],c='red',label='Exness')
axes[0].plot(data_init['Close'],c='dodgerblue',label='Init')
axes[0].scatter(data_exness[data_exness['spread'] >= 0.0001].index, data_exness['Close'][data_exness['spread'] >= 0.0001], s=300)
axes[0].set_ylabel('Price')
axes[0].legend()
axes[0].set_title('EURUSD_H4',fontsize = 16)

axes[1].plot(data_exness['spread'],c='coral')
axes[1].axhline(0.0001,color = 'blue', linestyle = 'dotted' ,linewidth =2)
axes[1].axhline(0,color = 'black', linestyle = '--' ,linewidth =2)
axes[1].axhline(-0.0001,color = 'blue', linestyle = 'dotted' ,linewidth =2)
axes[1].grid(False)
axes[1].set_yticks([0.0001,0,-0.0001])
axes[1].set_ylabel('Spread')
```
