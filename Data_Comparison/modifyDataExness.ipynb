```python
import pandas as pd
```

```python
data_init = pd.read_csv('../Data_Forex/EURUSD/EURUSD_H4 - EURUSD_H4.csv')
data_init
```

```python
data_exness = pd.read_csv('../Data_Forex/EURUSD/EURUSDm240.csv')
data_exness
```

```python
columns = ['Date','Time', 'Open', 'High','Low','Close','Volume']
data_exness.columns = columns
data_exness['DateTime'] = data_exness['Date'] + ' and ' + data_exness['Time']
data_exness['DateTime'] = pd.to_datetime(data_exness['DateTime'], format='%Y.%m.%d and %H:%M',errors='coerce')
data_exness['DateTime'] = data_exness['DateTime'].dt.strftime('%Y-%m-%d %H:%M:%S')
data_exness = data_exness.drop('Time', axis=1)
data_exness = data_exness.drop('Date', axis=1)
data_exness = data_exness.rename(columns={'DateTime': 'Time'})
data_exness
```

```python
columns = data_exness.columns.tolist()

# Specify the current position and target position of the column
current_position = 5  # Index of the current column position (zero-based)
target_position = 0  # Index where the column will be moved to

# Reorder the column list
columns.insert(target_position, columns.pop(current_position))
data_exness = data_exness[columns]
data_exness
```
